```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import gym
import dataset #my file
import vae #my file
import matplotlib.pyplot as plt
import vizdoomgym
import os
import tensorflow as tf
```

```python
# Specify which GPU(s) to use
os.environ["CUDA_VISIBLE_DEVICES"] = "1" 

frame_shape = [64,64]
env = gym.make('VizdoomTakeCover-v0')
dataset = dataset.Dataset(env,frame_shape)
vae = vae.VAE(frame_shape,dataset)
vae.load_json()


dataset.create_new_dataset(temporary = True)
```

```python
choosed_img = dataset.dataset[180]
imgplot = plt.imshow(choosed_img)
print("Real frame:")
plt.show()
```

```python
imgplot = plt.imshow(vae.synthesize_image(choosed_img))
print("Reconstructed frame:")
plt.show()
```
